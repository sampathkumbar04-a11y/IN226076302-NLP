# Task 1: Conceptual Understanding

## 1. Difference between "Love" and "love" in NLP



*   In Natural Language Processing (NLP), "Love" and "love" are considered different tokens when the text is case-sensitive.
*  This can unnecessarily increase the vocabulary size and may negatively impact the model’s efficiency and performance.


*  To handle this, text is usually converted to lowercase during preprocessing so that both "Love" and "love" are treated as the same word.




## 2. What happens if stopwords are not removed?

- The dataset becomes noisy because frequently used words like "is", "the", and "and" add little meaningful information.  

- Model performance may decrease, as these common words reduce the importance of more relevant keywords.  

- Computation time increases due to the presence of many unnecessary words in the data.  

- It becomes harder for the model to focus on important and meaningful information.



## 3. Scenarios where removing stopwords can be harmful

- In sentiment analysis, removing words like "not" can completely change the meaning of a sentence.
Example: "I am not happy" → removing "not" turns it into "I am happy", which is incorrect.

- In question answering systems, removing words like "what", "when", and "where" can remove the intent of the question.
Example: "What is your name?" becomes unclear and loses its purpose without "what".


## 4. Difference between Stemming and Lemmatization

- Stemming reduces words to their root form by simply trimming characters, which may not always result in meaningful words.
Example: "running" → "runn"

- Lemmatization converts words to their correct base or dictionary form (lemma), producing meaningful results.
Example: "running" → "run"

- Stemming is faster but less accurate.

- Lemmatization is slower but more accurate and linguistically meaningful.

# Task 2: Build Advanced Preprocessing Function
This function cleans raw text by removing noise like numbers, URLs, repeated characters, and unnecessary words.

```python
import re

def preprocess_text(text):
    # Handle empty or invalid input
    if not isinstance(text, str) or text.strip() == "":
        return [], ""

    # Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)

    # Remove email patterns
    text = re.sub(r"\S+@\S+", "", text)

    # Remove numbers
    text = re.sub(r"\d+", "", text)

    # Convert to lowercase
    text = text.lower()

    # Handle repeated characters (soooo → so)
    text = re.sub(r"(.)\1{2,}", r"\1", text)

    # Remove special characters and emojis
    text = re.sub(r"[^\w\s]", "", text)

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    # Tokenization
    tokens = text.split()

    # Remove short words (<=2) except "no" and "not"
    tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]

    # Join tokens back to sentence
    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence
    # Sample test sentences
test_sentences = [
    "I have 2 dogs",
    "This is    good",
    "soooo goooood!!!",
    "WOW!!! This is GREAT!!!",
    "Visit http://example.com now",
    "Email me at test@gmail.com",
    "Nooooo this is baaad!!!",
    "OK I am not happy",
]

# Run preprocessing
for sentence in test_sentences:
    tokens, cleaned = preprocess_text(sentence)

    print("Original Text   :", sentence)
    print("Cleaned Tokens  :", tokens)
    print("Cleaned Sentence:", cleaned)
    print("-" * 50)
```

# Task 3: Stress Testing

In this task, we test the preprocessing function on diverse real-world sentences including emojis, numbers, slang, repeated characters, URLs, and mixed case text.

```python
import re

# Preprocessing Function
def preprocess_text(text):
    if not isinstance(text, str) or text.strip() == "":
        return [], ""

    # Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)

    # Remove email patterns
    text = re.sub(r"\S+@\S+", "", text)

    # Remove numbers
    text = re.sub(r"\d+", "", text)

    # Convert to lowercase
    text = text.lower()

    # Handle repeated characters (soooo → so)
    text = re.sub(r"(.)\1{2,}", r"\1", text)

    # Remove special characters and emojis
    text = re.sub(r"[^\w\s]", "", text)

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    # Tokenization
    tokens = text.split()

    # Remove short words (<=2) except "no", "not"
    tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]

    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence


# Stress Test Sentences
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

# Run Testing
for sentence in test_sentences:
    tokens, cleaned = preprocess_text(sentence)

    print("Original Text   :", sentence)
    print("Cleaned Tokens  :", tokens)
    print("Cleaned Sentence:", cleaned)
    print("-" * 50)
```

# Task 4: Token Analytics

In this task, we analyze the processed tokens for each sentence by calculating:
- Total number of tokens
- Number of unique tokens
- Average token length

```python
# Task 4: Token Analytics (Code Only)

# Token Analysis Function
def analyze_tokens(tokens):
    if len(tokens) == 0:
        return 0, 0, 0

    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(word) for word in tokens) / total

    return total, unique, round(avg_len, 2)


# Run Analysis
max_noise_sentence = ""
max_noise_score = -1

best_sentence = ""
best_token_count = -1

for sentence in test_sentences:
    tokens, cleaned = preprocess_text(sentence)
    total, unique, avg_len = analyze_tokens(tokens)

    print("Sentence           :", sentence)
    print("Tokens             :", tokens)
    print("Total Tokens       :", total)
    print("Unique Tokens      :", unique)
    print("Avg Token Length   :", avg_len)
    print("-" * 50)

    # Noise estimation
    noise_score = len(sentence) - len(cleaned)
    if noise_score > max_noise_score:
        max_noise_score = noise_score
        max_noise_sentence = sentence

    # Most meaningful sentence
    if total > best_token_count:
        best_token_count = total
        best_sentence = sentence


# Final Results
print("\nMost Noisy Sentence      :", max_noise_sentence)
print("Most Meaningful Sentence :", best_sentence)
```

# Task 5: Frequency Analysis

In this task, we analyze the frequency of words by combining all tokens from the processed sentences.

We use the Counter class from the collections module to count how many times each word appears.

```python
from collections import Counter

tokens = ["python", "code", "data", "python", "analysis", "model", "data", "learning", "python", "model"]

# Count frequency
freq = Counter(tokens)

# Top 10 most frequent words
top_10 = freq.most_common(10)

# Top 5 least frequent words
least_5 = sorted(freq.items(), key=lambda x: x[1])[:5]

print("Top 10 Frequent Words:", top_10)
print("Top 5 Least Frequent Words:", least_5)
```

# Task 6: Build Full Pipeline

This function processes a list of text inputs and returns all tokens and cleaned sentences.

```python
import re

# Clean text function
def clean_text(sentence):
    # Convert to lowercase
    sentence = sentence.lower()
    # Remove punctuation
    sentence = re.sub(r'[^\w\s]', '', sentence)
    return sentence

# Tokenization function
def tokenize(sentence):
    return sentence.split()

# Full pipeline function
def full_pipeline(text_list):
    clean_sentences = []
    all_tokens = []

    for sentence in text_list:
        cleaned = clean_text(sentence)
        clean_sentences.append(cleaned)

        tokens = tokenize(cleaned)
        all_tokens.extend(tokens)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

# New Example input
text_list = [
    "Python is AWESOME!!!",
    "I love Data Science 💡",
    "Machine Learning is the FUTURE!!!",
    "Hello World 123!!!",
    "AI is changing the world!!!"
]

# Run pipeline
output = full_pipeline(text_list)

# Print output
print("Tokens:")
print(output["tokens"])

print("\nClean Sentences:")
print(output["clean_sentences"])
```
